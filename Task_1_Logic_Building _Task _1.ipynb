## 2. Pass / Fail Analyzer


```python
#Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
results=[]  #this is given to save a output in list for further steps

for mark in marks:
  if mark >=50:
     results.append("Pass")
  else:
    results.append("Fail")

print(results)
```

```python
print("Pass:", results.count("Pass"))
print("Fail:", results.count("Fail"))
```

## 3. Simple Data Cleaner


```python
#Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]

cleaned = [name.strip() for name in names] #strip function remove the leading and ending space

print(cleaned)
```

```python
lower_case=[name.lower() for name in cleaned] #lower function is used to make lowercase all letters
lower_case
```

## 4.Message Length Analyzer

```python
#Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

length =[len(message) for message in messages] #here to find length len() function is used
length
```

```python
for l in length:
    if l > 10:
        print("Flag")
    else:
        print("OK")
```

## 5. Error Message Detector

```python
#Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR") #by count() we can simply count output 

print("Total ERROR count:", error_count)
```
