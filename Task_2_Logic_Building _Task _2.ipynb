## Problem 1: Unique Words in a Sentence

```python
#Problem Statement
#Given a sentence, identify all unique words and display:
#The count of unique words
#The unique words themselves

sentence = input("Enter a sentence: ")

words = sentence.split()        # Convert sentence into list of words
unique_words = []  #to save the result or unique words

for word in words:
    if word not in unique_words:  #see for the remove duplicates the not in is used
        unique_words.append(word)  #append will add the output in unuique word list

count = len(unique_words)     # Count unique words

print("unique words count:", count)
print("unique words:",unique_words)
```

## Problem 2: Highest Salary from Employee Data


```python
#Problem Statement
#Store employee names and their salaries, and determine which employee has the highest salary.

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_salary = max(employees.values())  # to find the highest salary it is used

for name, salary in employees.items(): # here .items() are used because it will written a pair key-value
    if salary == highest_salary:
        print("Highest Salary:", name, "-", salary)

```

## Problem 3: Find Maximum and Minimum Values


```python
#Problem Statement
#Find the maximum and minimum values from a list of numbers.

List =[45, 22, 89, 10, 66]  #list is defined

max=List[0] # set a value to compare
min=List[0]

for i in List:   #loop is used to comapare each element
    if  i>max :  
        max=i
    if i<min:
        min=i
print("List :" ,List)
print("MAX :" ,max)
print("MIN :",min)
```

## Problem 4: Count Products Above a Price Threshold

```python
#Problem Statement
#Count how many products have a price greater than 1000.

prices = [450, 1200, 899, 1500, 300] #

count_price=[]

for price in prices:
    if price >1000:
        count_price.append(price)

print("Products above 1000:",len(count_price))
```

## Problem 5: Calculate Attendance Percentage

```python
#Problem Statement
#Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

attendance = ["P", "P", "A", "P", "P"]

total_days =len(attendance) # it give total number of days

present_days=attendance.count("P") #it count how many days presnt

percentage=(present_days/total_days *100) # formula to find percentage

print("Attendance Percentage :",percentage)
```

## Problem 7: Count Character Frequency

```python
#Problem Statement
#Count how many times each character appears in a given string.

password ="ppython" #INPUT is given

count_word ={}  #this will store the output in dictionary

for word in password:   #here logic is if a word in count_word it will increment by 1
    if word in count_word:
        count_word[word] +=1
    else:
        count_word[word] =1

print(count_word)
```
