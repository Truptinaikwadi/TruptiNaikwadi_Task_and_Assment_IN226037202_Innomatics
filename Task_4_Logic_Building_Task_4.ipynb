## Problem 1: Mobile Recharge Validation System


```python
def validate_recharge():  
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: â‚¹"))

        # Rule 1: Minimum amount check
        if amount < 50:
            print("Recharge amount must be at least â‚¹50.")
            print("Please try again")
            continue

        # Rule 2: Valid plan check
        if amount in valid_plans:
            print("Recharge successful")
            break
        else:
            print("Invalid plan selected.")
            print("Available plans are:", valid_plans)
            print("Please try again")

# Call the function
validate_recharge()
```

## Problem 2: Inventory Reorder Alert System


```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        
        if quantity < 15:
            print(product, ": Reorder Alert  (Stock =", quantity, ")")
        else:
            print(product, ": Stock OK (Stock =", quantity, ")")


inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 5,
    "Salt": 20
}

check_inventory(inventory)
```

## Problem 3: Student Result Processing System


```python
def process_result(marks_list):
    total = 0
    
  
    for mark in marks_list:
        total = total + mark
    
    average = total / len(marks_list)
    
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 45, 70, 55]


result = process_result(marks)
print("Result:", result)
```

## Problem 5: Employee Attendance Eligibility Checker


```python
def check_attendance(attendance_list):
    present_count = 0
    
  
    for status in attendance_list:
        if status == "P":
            present_count = present_count + 1
    
    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100
    
    print("Attendance Percentage:", percentage, "%")
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"



attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]


result = check_attendance(attendance)
print("Status:", result)
```

## Problem 6: Password Strength Checker


```python
def check_password(password):
    digit = False
    special = False

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    
    for char in password:
        if char.isdigit():
            digit = True
        
        if char in "@#$":
            special = True

  
    if digit and special:
        return "Strong Password "
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"


# Example
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
