## Problem Statement 1: Employee Performance Bonus Eligibility


```python
#dentify the highest performance score.
#Handle ties if multiple employees have the same highest score.
#Display all employees eligible for the top performance bonus.

#input givn
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

#highest score identified
highest_score =max(employees.values())

top_performers=[]

#the logic for loop is used to handle mulitple ties
for name,score in  employees.items():
     if score ==highest_score:
          top_performers.append(name)

#display of name 
print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      "(Score:", highest_score, ")")
```

## Problem Statement 2: Search Query Keyword Analysis


```python
#Convert the input to lowercase.
#Ignore common punctuation.
#Count the frequency of each keyword.
#Display only keywords searched more than once.

query = "Buy mobile phone buy phone online"

#convert to lowercase
query = query.lower()

#split into words
words = query.split()

#create empty dictionary
frequency = {}

# count words
for word in words:
    if word in frequency:
        frequency[word] = frequency[word] + 1
    else:
        frequency[word] = 1

#print words that appear more than once
result = {}

for word in frequency:
    if frequency[word] > 1:
        result[word] = frequency[word]

print(result)



```

## Problem Statement 3: Sensor Data Validation

```python
#Identify readings that are even numbers (valid readings).
#Store them as (hour_index, reading_value) pairs.
#Ignore odd readings (invalid readings).


sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem Statement 4: Email Domain Usage Analysis

```python
#Count how many users belong to each email domain.
#Calculate the percentage usage of each domain.

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count each domain
for email in emails:
    parts = email.split("@")
    domain = parts[1]
    
    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

# Calculate percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total_users) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

## Problem Statement 5: Sales Spike Detection


```python
#Calculate the average daily sales.
#Detect days where sales are more than 30% above average.
#Display the day number and sale value.

sales = [1200, 1500, 900, 2200, 1400, 3000]
# to find total
total = 0
for s in sales:
    total = total + s

#it will calculate total
average = total / len(sales)

# to find this limit is used
limit = average + (0.3 * average)

#Check each day
for i in range(len(sales)):
    if sales[i] > limit:
        print("Day", i + 1, ":", sales[i])
```

## Problem Statement 6: Duplicate User ID Detection

```python
#Identify duplicate user IDs.
#Display how many times each duplicate appears.

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_dict = {}

#count each user ID
for user in user_ids:
    if user in count_dict:
        count_dict[user] = count_dict[user] + 1
    else:
        count_dict[user] = 1

#it will print only duplicates
for user in count_dict:
    if count_dict[user] > 1:
        print(user, "→", count_dict[user], "times")
```
